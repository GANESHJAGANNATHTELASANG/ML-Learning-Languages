```python
class std:
    
    storage_type = "SSD"
    
    def __init__(self, name, RAM):
        self.name = name
        self.RAM = RAM

    def get_info(self):
        print(f"the laptop name : {self.name} and the RAM : {self.RAM}, storage type : {self.storage_type}")

lap1 = std("macOS", "16")
lap2 = std("lenova", "32")

lap1.get_info()
```

```python
#class method = as we use self, for this we use cls 
#class can only call class attributes it can't get the instance values or attributes
```

```python
class std:
    
    storage_type = "SSD"
    
    def __init__(self, name, RAM):
        self.name = name
        self.RAM = RAM

    @classmethod # we call it decorator
    def get_info(cls):
        print(f"the storage type is {cls.storage_type}")

lap1 = std("macOS", "16")
lap2 = std("lenova", "32")

std.get_info()
lap1.get_info()
```

```python
#static method = lke othere two we don;t want to send any parameters like cls, self there is nothing 
#it can't access any attributes like class and instance 
#it has decorator called @staticmethod
#we use it bcz ehich does not depend on any othe instance and class attributes which are usefull for class for that we initialise static
```

```python
class std:
    
    storage_type = "SSD"
    
    def __init__(self, name, RAM):
        self.name = name
        self.RAM = RAM

    @classmethod # we call it decorator
    def get_info(cls):
        print(f"the storage type is {cls.storage_type}")

    @staticmethod
    def disccount_price(price, dis):
        dis_pri= price - (dis * price / 100)
        print(f"the final price is {dis_pri}")
        

lap1 = std("macOS", "16")
lap2 = std("lenova", "32")

lap1.disccount_price(40_000, 10)
```

```python
# ==========================================
# 4 MAIN PILLARS OF OOP
# ==========================================

# 1. Encapsulation
# Wrapping data and methods into a single class and protecting the data.

# 2. Inheritance
# Acquiring the properties and methods of one class into another class.

# 3. Abstraction
# Hiding implementation details and showing only essential features.

# 4. Polymorphism
# One interface with many forms; the same method behaves differently for different objects.
```

```python
# ==========================================
#           ENCAPSULATION IN OOP
# ==========================================

# Definition:
# Encapsulation is the process of wrapping data (variables)
# and methods (functions) into a single class and controlling
# access to the data.

# Simple Meaning:
# - Data and methods are packed together inside a class.
# - Important data is protected from direct access or modification.

# Real-Life Example:
# ATM Machine
# - You can check balance.
# - You can withdraw money.
# - You can deposit money.
# - You cannot directly change your account balance.
#   You must use the ATM's functions.

# Why Do We Need Encapsulation?
# - Protects data from unauthorized access.
# - Hides internal implementation details.
# - Improves code organization.
# - Makes code easier to maintain.

# ==========================================
# Access Modifiers in Python
# ==========================================

# 1. Public
# - Accessible from anywhere.
# - Syntax: variable_name

# 2. Protected
# - Indicated by a single underscore (_variable_name).
# - Intended for internal use and inheritance.
# - Can still be accessed from outside (by convention).

# 3. Private
# - Indicated by double underscores (__variable_name).
# - Cannot be accessed directly from outside the class.
# - Accessed using Getter and Setter methods.

# ==========================================
# Getter
# ==========================================
# A Getter is a method used to read/access the value
# of a private variable.

# ==========================================
# Setter
# ==========================================
# A Setter is a method used to update/modify the value
# of a private variable.

# ==========================================
# Advantages of Encapsulation
# ==========================================
# 1. Data Security
# 2. Data Hiding
# 3. Better Code Organization
# 4. Easy Maintenance
# 5. Controlled Access to Data

# ==========================================
# Interview Definition
# ==========================================
# Encapsulation is the process of wrapping data and methods
# into a single class while controlling access to the data.
```

```python
#Best Practice: Use Getter and Setter 
```

```python
class std:
    def __init__(self, name):
        self.__name = name

    def get_name(self):
        return self.__name

    def set_name(self, newname):
         self.__name = newname


s1 = std("ganesh")
s1.get_name()
print("before setter function")
s1.set_name("rachana")
print("after setter function")
s1.get_name()
```

```python
# single inheritance 
```

```python
class teacher:
    def __init__(self, role):
        self.role = role

class student(teacher):
    def __init__(self, role, name):
        super().__init__(role)
        self.name = name
        

a1 = student("it is student", "ganesh")

print(a1.role, a1.name)
```

```python
#multilevel inheritance
```

```python
class teacher:
    def __init__(self, role):
        self.role = role

class student(teacher):
    def __init__(self, name):
        self.name = name

class child(student):
    def __init__(self, role, name, age):
        super().__init__(name)
        teacher.__init__(self, role)
        self.age = age

a = child("teacher", "ganesh master", "20")
print(a.role, a.name, a.age)
        
```

```python
#multiple inheritance
```

```python
class teacher:
    def __init__(self, role):
        self.role = role
class student:
    def __init__(self, name):
        self.name = name
class ta(teacher, student):
    def __init__(self, role, name, age):
        super().__init__(role)
        student.__init__(self, name)
        self.age = age
a = ta("tecaher", "ganesh", "20")

print(a.role, a.name, a.age)
```

```python
# ==========================================
# POLYMORPHISM
# ==========================================

# Definition:
# Polymorphism means "One Interface, Many Forms".
# The same method or operator behaves differently
# based on the object or data type.

# Types of Polymorphism:
# 1. Operator Overloading
# 2. Method Overriding
# 3. Duck Typing

# Operator Overloading:
# Same operator performs different operations
# depending on the data type.

# Method Overriding:
# A child class provides its own implementation
# of a method already defined in the parent class.

# Duck Typing:
# Python checks whether an object has the required
# method, not what type the object is.
```

```python
# type = 2. Method Overriding (Runtime Polymorphism)
#      = When a child class provides its own implementation of a method that already exists 
#        in the parent class, it is called Method Overriding.
```

```python
class Animal:
    def sound(self):
        print("all animals will sound's")
class Dog(Animal):
    def sound(self):
        print("all dogs barks")
class Cat(Animal):
    def sound(self):
        print("all cat's are meow's")
        
a = Animal()
b = Cat()
c = Cat()

b.sound()
c.sound()
a.sound()
```
