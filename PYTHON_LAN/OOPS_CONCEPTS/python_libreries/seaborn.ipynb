```python
tips = tips_data.copy()
```

```python
tips
```

```python
sns.scatterplot(tips, x="total_bill", y="tip", hue="sex", style="smoker", size="size")
```

```python
sns.relplot(tips, x="total_bill", y="tip", hue="sex", style="smoker", size="size",col="day" ,kind="scatter")
```

```python
sns.relplot(tips, x="total_bill", y="tip", hue="sex", style="smoker", size="size",row="day" ,kind="scatter")
```

```python
sns.relplot(tips, x="total_bill", y="tip", hue="sex", style="smoker", size="size",col="day", col_wrap=2 ,kind="scatter")
```
