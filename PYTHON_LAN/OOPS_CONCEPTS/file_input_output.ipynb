```python
try:
    num = int(input("enter the number to divide"))
    an = 10/num
except ValueError:
    print("not valid one")
except ZeroDivisionError:
    print("can't divid by zero")
else:
    print(f"the ans is {an}")
```

```python
sq = [ i*i for i in range(7) if i != 0]
print(sq)
```

```python
nu = [-1, 4, -9, 16, -25, 36]
num = [i for i in nu if i<0 ]
print(num)
```

```python
nu = [-1, 4, -9, 16, -25, 36]
num = [0 if i<0 else i for i in nu]
print(num)
```
