```python
class Student:

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

s1 = Student("Ganesh", 95)
s2 = Student("Rahul", 88)

print(s1.name, s1.marks)
print(s2.name, s2.marks)
```

```python
class std:
    
    storage_type = "SSD"
    
    def __init__(self, name, RAM):
        self.name = name
        self.RAM = RAM

    def get_info(self):
        print(f"the laptop name : {self.name} and the RAM : {self.RAM}, storage type : {self.storage_type}")

lap1 = std("macOS", "16")
lap2 = std("lenova", "32")

lap1.get_info()
```

```python
#class method = as we use self, for this we use cls 
#class can only call class attributes it can't get the instance values or attributes
```

```python
class std:
    
    storage_type = "SSD"
    
    def __init__(self, name, RAM):
        self.name = name
        self.RAM = RAM

    @classmethod # we call it decorator
    def get_info(cls):
        print(f"the storage type is {cls.storage_type}")

lap1 = std("macOS", "16")
lap2 = std("lenova", "32")

std.get_info()
lap1.get_info()
```

```python
#static method = lke othere two we don;t want to send any parameters like cls, self there is nothing 
#it can't access any attributes like class and instance 
#it has decorator called @staticmethod
#we use it bcz ehich does not depend on any othe instance and class attributes which are usefull for class for that we initialise static
```

```python
class std:
    
    storage_type = "SSD"
    
    def __init__(self, name, RAM):
        self.name = name
        self.RAM = RAM

    @classmethod # we call it decorator
    def get_info(cls):
        print(f"the storage type is {cls.storage_type}")

    @staticmethod
    def disccount_price(price, dis):
        dis_pri= price - (dis * price / 100)
        print(f"the final price is {dis_pri}")
        

lap1 = std("macOS", "16")
lap2 = std("lenova", "32")

lap1.disccount_price(40_000, 10)
```
