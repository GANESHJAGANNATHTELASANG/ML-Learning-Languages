```python
#2. We can use operators like and, or ,  not, !=, == ,>= ,<= , , etc.
```

```python
dfe.query("AGE > 29 & INCOME >  50000")
```

```python
salary = 50000
dfe.query("AGE > 29 and INCOME >  @salary")
```

```python
salary = 50000
dfe.query("AGE > 29 or INCOME >  @salary")
```

```python
salary = 50000
dfe.query("AGE > 19 == INCOME >  @salary")
```

```python
dfe.query("AGE == 29")
```

```python
va = "USA"
dfe.query("AGE != 29 and COUNTRY == @va")
```
